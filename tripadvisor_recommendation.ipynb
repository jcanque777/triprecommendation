# Using NLP on Comments

```python
# first of all, let us remove some useless words to see how our documents change
stop_words = set(stopwords.words('english')) 
```

```python
#set index- will be needed later for calling the matrix
df.set_index('name',inplace=True)
```

```python
# calculate cosine similarity between documents - we use tfidf
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
```

```python
#create the matrix
tfidf_matrix = tf.fit_transform(df['comments_real2'])
```

```python
#use consine similarity to get values for matrix
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
#match index
indices = pd.Series(df.index)
```

```python
indices[indices == "Central Park"].index[0]
```

```python
from copy import deepcopy
```

```python
def recommendations(name, similarity_matrix = similarity_matrix, df=df):
    
    recommended_activities = []
    
    # gettin the index of the activity that matches the name
    idx = indices[indices == name].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(similarity_matrix[idx]).sort_values(ascending = False)

    # getting the indexes of the 25 most similar activities except itself
    top_25_indexes = list(score_series.iloc[1:26].index)
    
    # populating the list with the names of the top 25 matching hotels
    # filters the names that are only in London
    
    temp = deepcopy(df)
    results = []
    for i in top_25_indexes:
        if temp.iloc[i, :].location_nyc == 0:
            results.append(i)
    
    for i in results:
        recommended_activities.append(list(df.index)[i])
                
    return recommended_activities
```

```python
recomendations = input("Where do you like to go in NYC?")
recommendations(recomendations)
```
